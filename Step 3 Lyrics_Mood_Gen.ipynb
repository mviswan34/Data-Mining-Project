```python
from transformers import AutoTokenizer, AutoModelWithLMHead

tokenizer = AutoTokenizer.from_pretrained("mrm8488/t5-base-finetuned-emotion")

model = AutoModelWithLMHead.from_pretrained("mrm8488/t5-base-finetuned-emotion")

def get_emotion(text):
  input_ids = tokenizer.encode(text + '</s>', return_tensors='pt')

  output = model.generate(input_ids=input_ids,
               max_length=2)
  
  dec = [tokenizer.decode(ids) for ids in output]
  label = dec[0]
  return label

df['label'] = df['cleaned_lyrics'].apply(get_emotion)


```

```python



tokenizer = BertTokenizer.from_pretrained('brunokreiner/lyrics-bert')
model = BertForSequenceClassification.from_pretrained('brunokreiner/lyrics-bert')
# Function to tokenize lyrics
def tokenize(lyrics):
    tokens = tokenizer(lyrics, truncation=True, padding=True, return_tensors='pt')
    return tokens

# Apply the cleaning function to the 'col2' containing lyrics
df['tokenized_lyrics'] = df['Cleaned Lyrics'].apply(tokenize)

# Save the cleaned DataFrame to a new CSV file
# df.to_csv('cleaned_lyrics.csv', index=False)
df
```

```python
def predict_emotion(tokens):
    with torch.no_grad():
        outputs = model(**tokens)
    return outputs['logits']

# Apply the tokenization and prediction functions to the 'col2' containing lyrics
df['emotion_logits'] = df['tokenized_lyrics'].apply(predict_emotion)

# Save the DataFrame with tokenized lyrics and emotion logits to a new CSV file
# df.to_csv('emotion_predictions.csv', index=False)
```
